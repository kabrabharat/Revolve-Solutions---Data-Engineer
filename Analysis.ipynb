```python
import pandas as pd
import json
import os
from datetime import datetime
```

```python
customer_df = pd.read_csv("./input_data/starter/customers.csv")
customer_df.head()
```

```python
product_df = pd.read_csv("./input_data/starter/products.csv")
product_df.head()
```

```python
transaction_dirs = os.listdir("./input_data/starter/transactions/")
transaction_dirs[:10]
```

```python
transaction_dates = []
for dirs in transaction_dirs:
    transaction_dates.append(datetime.strptime(dirs.split("=")[-1], '%Y-%m-%d'))
transaction_dates[:10]
```

```python
type(transaction_dates[0])
```

```python
transaction_dates[3].isoweekday()
```

```python
weekly_transactions = {}
weekly_list = []
flag = 0
for tr_date_index in range(len(transaction_dates)):
    
    if flag == 0:
        start_date = transaction_dirs[tr_date_index].split('=')[-1]
        flag = 1
        weekly_list = []
    
    weekly_list.append(transaction_dirs[tr_date_index])
    
    if transaction_dates[tr_date_index].isoweekday() == 7:
        end_date = transaction_dirs[tr_date_index].split('=')[-1]
        weekly_transactions[start_date + " to " +end_date] = weekly_list
        flag = 0
        
```

```python
weekly_transactions
```

```python
f = open("./input_data/starter/transactions/d=2018-12-01/transactions.json")
f.read(200)
```

```python
weekly_wise_transactions = {}
one_week_transactions = []
for date_range, dates in weekly_transactions.items():
    
    for each_tr_date in dates:
        
        transaction_file = open("./input_data/starter/transactions/"+ str(each_tr_date) +"/transactions.json", 'r')

        for line in transaction_file:
            transactions_data = json.loads(line)
            one_week_transactions.append(transactions_data)

    weekly_wise_transactions[str(date_range)] = one_week_transactions
transaction_file.close()
```

```python
weekly_wise_transactions
```

```python
df = pd.DataFrame()
for date_range, transactions in weekly_wise_transactions.items():
    for transaction in transactions:
        df_new = pd.DataFrame(transaction)
        df = pd.concat([df, df_new])
    break
df
```
